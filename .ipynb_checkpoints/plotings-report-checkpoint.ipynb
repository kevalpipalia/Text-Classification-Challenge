```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(context="notebook", palette=("husl"))
```

```python
comparison_grid = {'without preprocessing': 0.77954, 'Stemming': 0.77495 , 'Lemmatizing':0.77825, 'Stopwords removal':0.76962, 'Final Text cleaning pipeline':0.77836}
plt.barh(list(comparison_grid.keys()), list(comparison_grid.values()))
plt.xlim(0.7,0.8)
plt.xlabel('Accuracy on validation set')
plt.ylabel('Preprocessing technique used')
plt.title('Effects of preprocessing techniques on accuracy')
```
